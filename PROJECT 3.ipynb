```python
###Now we have to create a model that should work for both the tracks.

log=[]
with open('C:/users/SIDAUS/Carnd-Behavioral-Cloning-P3-master/DATA/driving_log.csv','r') as csvfile:
    logname=csv.reader(csvfile)
    for logs in logname:
        log.append(logs)

training_samples, validation_samples=train_test_split(log,test_size=0.25)

def generate(samples,batch_size=32):
    value=40
    NUM=len(samples)
    while 1:                                            ###infinite loop
        shuffle(samples)
        for offset in range(0,NUM,batch_size):
            batch_samples=samples[offset:offset+batch_size]

            images=[]
            labels=[]
            for batch in batch_samples:                         ###adding the flipped images
                path = batch[0]
                img=cv2.cvtColor(cv2.imread(path),cv2.COLOR_BGR2RGB)+value
                images.append(cv2.flip(img,1))
                label=float(batch[3])*-1
                labels.append(label)        

            correction_factor=0.2276                 ###randomly generated number
            for batch in batch_samples:
                for i in range (3):
                    path = batch[i]
                    img=cv2.cvtColor(cv2.imread(path),cv2.COLOR_BGR2RGB)+value
                    images.append(img)
                    if i==1:                         #### left image
                        label=float(batch[3])+correction_factor
                    elif i==2:                       #### right image
                        label=float(batch[3])-correction_factor
                    else:                            #### centre image
                        label=float(batch[3])
                    labels.append(label)   
            
            x_train=np.array(images)
            y_train=np.array(labels)
            
            yield shuffle(x_train,y_train)
            
###generating the required data

Train=generate(training_samples,batch_size=32)
Valid=generate(validation_samples,batch_size=32)
```

```python
###Using modified NVIDIA network (most of the changes were random and instinctive and someother changes were suggested by fellow classmates at slack)
model=Sequential([
        Lambda(lambda x:x/127.5-1, input_shape=(160,320,3)),
        Cropping2D(cropping=((50,25),(0,0))),                     ###output size = (85,320,3)
        Conv2D(24,5,5,subsample=(2,2),activation='relu'),  ###output size = (41,158,24)
        BatchNormalization(),
        #Dropout(0.25),
        Conv2D(36,5,5,subsample=(2,2),activation='relu'),  ###output size = (19,78,36)
        BatchNormalization(),
        #Dropout(0.25),
        Conv2D(48,5,5,subsample=(1,1),activation='relu'),  ###output size = (14,73,48)
        BatchNormalization(),
        #Dropout(0.25),
        Conv2D(64,3,3,activation='relu'),                  ###output size = (4,24,64)
        BatchNormalization(),
        #Dropout(0.25),
        Conv2D(64,3,3,subsample=(1,1),activation='relu'),  ###output size = (1,22,64)      
        #Dropout(0.25),
        Flatten(),
        Dense(100),
        Dropout(0.25),
        Dense(50),
        Dense(10),
        Dense(1)
])
```

```python
###Lets compile it
model.compile(loss='mse',optimizer='adam')
model.fit_generator(Train,samples_per_epoch=len(training_samples),
                   validation_data=Valid,nb_val_samples=len(validation_samples),
                   nb_epoch=10)


model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)

model.save('model.h5')
```
